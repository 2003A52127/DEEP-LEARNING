```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
plt.rcParams.update()
x = np.arange(-1,1,0.01)
h11 = 1/(1+np.exp(-((500*x)+30)))
h12 = 1/(1+np.exp(-((500*x)-30)))
h21=h11-h12
```

```python
plt.title("GRAPH FOR THE FUNCTION h11(x)")
plt.plot(x,h11)
plt.ylabel("$h_{11}(x)$")
plt.show()
```

```python
plt.title("GRAPH FOR THE FUNCTION h12(x")
plt.plot(x,h12,"r-")
plt.ylabel("$h_{11}(x)$")
plt.show()
```

```python
plt.title("GRAPH FOR THE FUNCTION h21(x)")
plt.ylabel("$h_{21}(x)$")
plt.plot(x,h21,"g-")
plt.show()
```

```python
def h11(x1,x2):
    return 1/(1 + np.exp(-(x1 + 50*x2 + 100)))

def h12(x1,x2):
    return 1/(1 + np.exp(-(x1 + 50*x2 - 100)))

def h13(x1,x2):
    return 1/(1 + np.exp(-(50*x1 + x2 + 100)))

def h14(x1,x2):
    return 1/(1 + np.exp(-(50*x1 + x2 - 100)))

def h21(x1,x2):
    return h11(x1,x2) - h12(x1,x2)

def h22(x1,x2):
    return h13(x1,x2) - h14(x1,x2)

def h31(x1,x2):
    return h21(x1,x2) + h22(x1,x2)

def f(x1,x2):
    return 1/(1 + np.exp(-(100*h31(x1,x2) - 200)))
x1 = np.arange(-5,5,0.001)
x2 = np.arange(-5,5,0.001)
X, Y = np.meshgrid(x1, x2)
```

```python
Z = h11(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("GRAPH FOR THE FUNCTION h11(x1,x2)")
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('h11');
```

```python
Z = h12(X, Y)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$h_{12}$');
ax.set_title("GRAPH FOR THE FUNCTION h12(x1,x2)")
```
